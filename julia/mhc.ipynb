## One hot encoding

````julia
codes = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L',
         'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']


function one_hot_encode(seq)
    o = list(set(codes) - set(seq))
    s = DataFrame(list(seq))    
    x = DataFrame(np.zeros((len(seq),len(o)),dtype=int),columns=o)    
    a = s[0].str.get_dummies(sep=',')
    a = a.join(x)
    a = a.sort_index(axis=1)    
    e = a.values.flatten()
    return e
end    
````
