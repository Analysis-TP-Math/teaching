```python
import glob,os,sys
import pandas as pd
import numpy as np
pd.set_option('display.width', 100)
pd.set_option('max_colwidth', 90)
from IPython.display import display, HTML
%matplotlib inline
import matplotlib as mpl
import pylab as plt
import seaborn as sns
sns.set_style("ticks", {'axes.facecolor': '#F7F7F7','axes.grid': False,'legend.frameon':True, 'legend.fontsize':12})
sns.set_context("notebook", font_scale=1.6)
mpl.rcParams['savefig.dpi'] = 90
```

```python
df=pd.read_csv('SH_seaice_extent_final_v2.csv',index_col=0)
print df[:3]
df=df[df.year>2005]
```

```python
#g = sns.factorplot(y='Extent',x='date', hue_order='year', data=df, col='year', kind="bar",                    
#                   col_wrap=2,size=4,aspect=1.5,sharex=False, legend_out=True)

groups = df.groupby('year')
pal = sns.color_palette("hls", n_colors=len(groups))
fig,axs=plt.subplots(1,1,figsize=(12,7))
i=0
for yr,g in groups:
    g.plot(y='Extent',x='date',lw=1.2,ax=axs,label=yr,c=pal[i])
    i+=1
```
